```python
import yfinance as yf
import mplfinance as mpf
import pandas as pd
```

```python
user_input1, user_input2 = input('Please enter 1st ticker:'), input('Please enter 2nd ticker:')

#user_input1
#user_input2
```

```python
ticker1 = user_input1
ticker2 = user_input2
```

```python
print(f'Selected 1st ticker: {ticker1}')
print(f'Selected 2nd ticker: {ticker2}')
```

```python
history_input1 = yf.Ticker(ticker1).history(period='3mo')
print(f'Selected ticker: {ticker1}')
print(f'3 months high: {history_input1.High.max().astype(int)}')
print(f'3 months low: {history_input1.Low.min().astype(int)}')
```

```python
history_input1.index
```

```python
history_input2 = yf.Ticker(ticker2).history(period='3mo')
print(f'Selected ticker: {ticker2}')
print(f'3 months high: {history_input2.High.max().astype(int)}')
print(f'3 months low: {history_input2.Low.min().astype(int)}')
```

You can also download two or more tickers simultaneously. Just pass the ticker as a list.
If you want to remove Adjust Close Price column then you can do so by setting auto_adjust= True
You can download dividends and splits of stock by setting actions =”inline”.


```python
history_input3 = yf.download([ticker1, ticker2], start='2023-01-01',
end='2023-02-21',
progress=False, auto_adjust=True, actions="inline")
history_input3.head(2)
```
